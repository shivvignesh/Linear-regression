```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
boston=load_boston()
```

```python
features=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
target=pd.DataFrame(boston.target,columns=['Target'])
```

```python
features
```

```python
target
```

```python
data=pd.concat([features,target],axis=1)
```

```python
data
```

```python
x=data['RM']
```

```python
y=data['Target']
```

```python
x=np.array((x-x.min())/(x.max()-x.min()))
```

```python
y=np.array((y-y.min())/(y.max()-y.min()))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.size
```

```python
def hypothesis(a,b,c,x):
    return a*np.power(x,2)+b*x+c
```

```python
def error(a,b,c,x,y):
    err=0
    m=len(x)
    
    for i in range(m):
        err+=np.power(hypothesis(a,b,c,x[i])-y[i],2)
        
    return (1/(2*m))*err
        
```

```python
def step_gradient(a,b,c,x,y,learning_rate):
    grad_a=0
    grad_b=0
    grad_c=0
    m=len(x)
    
    for i in range(m):
        grad_a+=(1/m)*(hypothesis(a,b,c,x[i])-y[i])*np.power(x[i],2)
        grad_b+=(1/m)*(hypothesis(a,b,c,x[i])-y[i])*x[i]
        grad_c+=(1/m)*(hypothesis(a,b,c,x[i])-y[i])
        
    a=a-grad_a*learning_rate
    b=b-grad_b*learning_rate
    c=c-grad_c*learning_rate
    
    return a,b,c
    
```

```python
def descend(init_a,init_b,init_c,x,y,learning_rate,iterations):
    a = init_a
    b = init_b
    c = init_c
    for i in range(iterations):
        e = error(a, b, c, x, y)
        if i % 1000 == 0:
            print(f"Error: {e}, a: {a}, b: {b}, c: {c}")
        
        a, b, c = step_gradient(a, b, c, x, y, learning_rate)

    return a, b, c
```

```python
a = 0.7456601990528254
b = 0.23471805515139219
c = 0.0494096579740407
learning_rate = 0.05
iterations = 20000
```

```python
final_a, final_b, final_c = descend(a, b, c, x_train, y_train, learning_rate, iterations)
```

```python
plt.plot(x_train,y_train,"r.",
         x_train,hypothesis(a,b,c,x_train),"g.",
         x_train,hypothesis(final_a,final_b,final_c,x_train),"b.")
```
