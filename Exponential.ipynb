```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
boston=load_boston()
```

```python
features=pd.DataFrame(boston.data,columns=boston.feature_names)
target=pd.DataFrame(boston.target,columns=['target'])
```

```python
data=pd.concat([features,target],axis=1)
```

```python
data
```

```python
x=data['RM']
y=data['target']
```

```python
x=np.array((x-x.min())/(x.max()-x.min()))
y=np.array((y-y.min())/(y.max()-y.min()))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.size
```

```python
def hypothesis(a,b,x):
    return np.exp(a*x+b)
```

```python
def error(a,b,x,y):
    err=0
    m=len(x)
    
    for i in range(m):
        err+=np.power(hypothesis(a,b,x[i])-y[i],2)
        
    return (1/(2*m))*err
```

```python
def step_gradient(a,b,x,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    
    for i in range(m):
        grad_a+=(1/m)*(hypothesis(a,b,x[i])-y[i])*np.exp(a*x[i]+b)*x[i]
        grad_b+=(1/m)*(hypothesis(a,b,x[i])-y[i])*np.exp(a*x[i]+b)
        
    a=a-grad_a*learning_rate
    b=b-grad_b*learning_rate
    
    return a,b
```

```python
def descend(init_a,init_b,x,y,learning_rate,iterations):
    a=init_a
    b=init_b 
    
    for i in range(iterations):
        e=error(a,b,x,y)
    
        if i%1000 == 0:
            print(f"Error:{np.sqrt(e)},a:{a},b:{b}")
    
        a,b=step_gradient(a,b,x,y,learning_rate)
    
    return a,b 
```

```python
a=0
b=0
learning_rate=0.05
iterations=10000
```

```python
final_a,final_b=descend(a,b,x_train,y_train,learning_rate,iterations)
```

```python
plt.plot(x_train,y_train,"r.",
         x_train,hypothesis(a,b,x_train),"g.",
         x_train,hypothesis(final_a,final_b,x_train),"b.")
```
